```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# 读取数据集,并且在datafram表结构中,命名为df
# 原始数据只有一列,以' '分割,这里读取数据的时候直接切分
df1 = pd.read_csv('used_car_1.csv', sep=' ')
df2 = pd.read_csv('used_car_2.csv', sep=' ')
df3 = pd.read_csv('used_car_3.csv', sep=' ')
```

```python
# 合并成一个数据集
df = pd.concat([df1, df2, df3])
```

```python
df
```

```python
# 去掉匿名特征
df = df.drop(['v_0', 'v_1', 'v_2', 'v_3', 'v_4', 'v_5', 'v_6', 'v_7', 'v_8', 'v_9',
             'v_10', 'v_11', 'v_12', 'v_13', 'v_14'], axis=1)
```

### 特征介绍 

Filed       Description<br>
SaleID      交易ID，唯一编码<br>
name       汽车交易名称，已脱敏<br>
regDate     汽车注册日期, 例如20160101<br>
model      车型编码<br>
brand      汽车品牌<br>
bodyType    车身类型:豪华轿车: 0,微型车: 1, 厢型车: 2, 大巴车: 3, 敞篷车: 4, 双门车: 5, 商务车: 6, 搅拌车: 7<br>
fuelType   燃油类型: 汽油: 0, 柴油: 1, 液化石油气: 2, 天然气: 3, 混合动力: 4, 其他: 5<br>
gearbox    变速箱: 手动: 0, 自动: 1<br>
power     发动机功率: 范围[0, 600]<br>
kilometer  汽车已行驶公里, 单位万km<br>
notRepairDamage  汽车有尚未修复的损坏: 是: 0, 否: 1<br>
regionCode  地区编码<br>
seller    销售方: 个体: 0, 非个体: 1<br>
offerType  报价类型: 提供: 0, 请求: 1<br>
creatDate  汽车上线时间, 即开始售卖时间<br>
price    二手车交易价格(预测目标)<br>
v系列特征 匿名特征,包含v0-14在内15个匿名特征<br>

### 数据分析 

```python
# 使用众数填充缺失值
bodyType_mode = df['bodyType'].mode()
fuelType_mode = df['fuelType'].mode()
gearbox_mode = df['gearbox'].mode()
df = df.fillna({'bodyType': 'bodyType_mode', 'fuelType': 'fuelType_mode', 'gearbox': 'gearbox_mode'})
```

```python
# 检查数据集是否有空缺值
df.isnull().sum()
```

```python
import seaborn as sns
# 汽车注册日期(regDate)和价格(price)关系图
sns.regplot(x='regDate', y='price', data=df)
```

```python
# 发动机功率(power)和价格(price)关系图
sns.regplot(x='power', y='price', data=df)
```

```python
# 汽车已行驶公里(kilometer)和价格(price)关系图
sns.regplot(x='kilometer',y='price', data=df)
```

```python
# 变速箱(gearbox)和价格(price)关系图
df.gearbox = df.gearbox.map(lambda x: x if type(x) != str else np.NaN)
sns.regplot(x='gearbox',y='price', data=df)
```

```python
# 车身类型(bodyType)和价格(price)关系图
df.bodyType = df.bodyType.map(lambda x: x if type(x) != str else np.NaN)
sns.regplot(x='bodyType', y='price', data=df)
```

```python
# 车型编码(model)和价格(price)关系图
sns.regplot(x='model',y='price', data=df)
```

```python
# 燃油类型(fuelType)和价格(price)关系图
df.fuelType = df.fuelType.map(lambda x: x if type(x) != str else np.NaN)
sns.regplot(x='fuelType',y='price',data=df)
```
